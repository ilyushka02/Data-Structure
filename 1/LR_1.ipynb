## Задание 1

Разработайте алгоритм, который запрашивает у пользователя четыре целых числа и затем определяет, есть ли среди введенных чисел одинаковые или нет.

Программа:

```python
from random import randint

def check_unique(
    list_numbers: list
)-> bool:
    """
    This function checks for duplicates in the list
    
    Return:
    
        True - duplicates finde
        
        False - duplicate not finde
    """

    return len(set(list_numbers)) != len(list_numbers)
    
# Size list
len_list = 4

# create list
input_list_numbers = [int(input()) for i in range(0, len_list)]

# check duplicates in list
print(f"You`re list: {input_list_numbers}")
print("Duplicates in you`re list: ", check_unique(input_list_numbers))
```

## Задание 2

Разработайте алгоритм, который запрашивает у пользователя четыре целых числа и затем находит третье по величине число, если оно существует.

Блок схема:



Программа:

```python
# import
from random import randint

# algorithms sort bubble
def sort(
    numbers: list
)->list:  
    """
    This function sort values in list

    Return:
    
        numbers: list
    """
    
    # Set True fore minimal one step
    swap = True
    while swap:
        swap = False
        
        # fore everyone elemenst check value
        for idx in range(len(numbers) - 1):
            
            # swap elemens
            if numbers[idx] > numbers[idx + 1]:
                numbers[idx], numbers[idx + 1] = numbers[idx + 1], numbers[idx]
                
                # again set True fore going
                swap = True
    return numbers

def search_third_max(
    list_numbers: list
)-> int:
    """
    This function serch thrid max numbers
    
    Returns:
    
        number: int
    """
    
    if len(list_numbers) < 3:
        raise Exception(f"""
            minimal size for list  equally 3
            size you're list  {len(list_numbers)}
        """)
    
    list_numbers = list(set(list_numbers))
    
    if len(set(list_numbers)) < 3:
        raise Exception(f"""
          expected three unique values 
          received {len(set(list_numbers))}
        """)

    number = sort(list_numbers)[-3]
    
    return number, list_numbers 
    
    
# Size list
len_list = 4

# create list
input_list_numbers = [int(input())  for i in range(0, len_list)]

print(f"You`re list: {input_list_numbers}")
output = search_third_max(input_list_numbers)
print(f"You`re sort list: {output[1]}")
print(f"Thrid max values in you`re list: {output[0]}", )
```

## Задание 3

Разработайте алгоритм, который случайным образом задает четыре целых числа и затем определяет максимальное и минимальное число, а также находит количество максимальных и минимальных чисел среди введенных. 

Блок схема:



Программа:

```python
# import
from random import randint

def search_min_and_max_value(
    list_numbers: list
) -> int:
    """
    This function serch max and min values in list.
    And counting values equally max values or min values
    
    Returns:
    
        max_value: int
        min_value: int
        count_max: int
        count_min: int
    """
    
    min_value = max_value = list_numbers[0]
    
    list_numbers_uneque = set(list_numbers)
    for num in list_numbers_uneque:
        
        min_value = num if min_value > num else min_value
        
        max_value = num if max_value < num else max_value
        
    count_max = len([num for num in list_numbers if num == max_value])
    count_min = len([num for num in list_numbers if num == min_value])
    return max_value, min_value, count_max, count_min


# Size list
len_list = 4

# create list
input_list_numbers = [randint(0, 9) for i in range(0, len_list)]

print(f"You`re list: {input_list_numbers}")
output = search_min_and_max_value(input_list_numbers)
print(f"""max value = {output[0]}\tcount max value = {output[2]}
min value = {output[1]}\tcount min value = {output[3]}
""")
```
