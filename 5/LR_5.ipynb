```python
from random import randint
from typing import Union
```

```python
class Stack:
    def __init__(self, max_size=None):
        self.stack = []
        self.size = max_size
        
    def clear(self):
        self.stack = []
        
    def is_empty(self):
        return len(self.stack) == 0
    
    def is_full(self):
        if self.size is None:
            return False
        else:
            return self.size == len(self.stack)  
    
    def push(self, item):
        if self.is_full() == False:
            self.stack.append(item)
        else:
            raise Exception("The index has gone abroad")
    
    def pop(self):
        if self.is_empty():
            return None
        removed = self.stack.pop()
        return removed
    
    def peek(self):
        if self.is_empty():
            return None
        removed = self.stack[-1]
        return removed
```

```python
s = Stack(3)
```

```python
for i in range(0, 2):
    s.push(randint(1, 10))
```
